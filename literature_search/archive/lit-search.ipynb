```python
def askGPT3(title, abstract):

    prompt = 'In the scientific article titled: "' + title + '", with the abstract "' + abstract + '",  can we determine that the article concerns production using synthetic communities, consortia or co-culture?'

    response = openai.Completion.create(
        model="text-davinci-002",
        prompt=prompt,
        temperature=0.0,
        max_tokens=193,
        top_p=0.5,
        frequency_penalty=0,
        presence_penalty=0
    )

    return response.choices[0].text 
```
