# Subpopulation percentage

Retrieving subpopulation percentage data from [Fig 5.](subpopulation_percentage/subpopulation_percentage.png) in RA paper.
Have used [WebPlotDigitizer](https://automeris.io/WebPlotDigitizer/) to get each dataframe with points.

```python
import pandas as pd

# load each df
sp_cal2 = pd.read_csv("subpopulation_percentage/CAL2.csv", header=None, names=["time", "subpopulation_percentage"])
sp_cal2["strain"] = "CAL2"
sp_sal9 = pd.read_csv("subpopulation_percentage/SAL9.csv", header=None, names=["time", "subpopulation_percentage"])
sp_sal9["strain"] = "SAL9"
sp_mam2 = pd.read_csv("subpopulation_percentage/MAM2.csv", header=None, names=["time", "subpopulation_percentage"])
sp_mam2["strain"] = "MAM2"

# concat into one df
df = pd.concat([sp_cal2, sp_mam2, sp_sal9], axis=0)

# save df
df.to_csv("subpopulation_percentage/data.csv", index=False)
```

```python
import seaborn as sns
sns.lineplot(df, x="time", y="subpopulation_percentage", hue="strain")
```

# Concentrations of products

Retrieving concentrations (mg/L) of products from [Fig 5.](concentrations/concentrations.png.png) in RA paper.
Have used [WebPlotDigitizer](https://automeris.io/WebPlotDigitizer/) to get each dataframe with points.

```python
import pandas as pd

# load each df
ca = pd.read_csv("concentrations/CA.csv", header=None, names=["time", "concentration"])
ca["product"] = "CA"
saa = pd.read_csv("concentrations/SAA.csv", header=None, names=["time", "concentration"])
saa["product"] = "SAA"
ra = pd.read_csv("concentrations/RA.csv", header=None, names=["time", "concentration"])
ra["product"] = "RA"

# concat into one df
df = pd.concat([ca, saa, ra], axis=0)

# save df
df.to_csv("concentrations/data.csv", index=False)
```

```python
import seaborn as sns
sns.lineplot(df, x="time", y="concentration", hue="product")
```
