```python
import pandas as pd
df4 = pd.read_csv("results/prod_env_yields_tri_xyl_glc_new_params.csv")
```

```python
import pandas as pd
import plotly.express as px

fig = px.scatter_3d(df4, x='time', y='biomass', z='RA', color='yield_factor')

marker_size = 4  # Adjust the size of the scatter points as needed

fig.update_traces(marker=dict(size=marker_size))

# Customize the layout and set the width and height of the figure
fig.update_layout(
    width=850,  # Set the width of the figure in pixels
    height=800    # Set the height of the figure in pixels
)


fig.write_image("results/figures/call11_sal11_mam3/yield_envelope.svg")

# Show the plot
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax =plt.subplots(1,2, figsize=(12, 4))
sns.lineplot(data=df4, x="time", y="biomass", hue="yield_factor", ax=ax[0], palette="flare")
sns.lineplot(data=df4, x="time", y="RA", hue="yield_factor", ax=ax[1], legend=False, palette="flare")

# Add labels, legends, etc. if needed
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Biomass (gDW)')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('RA produced (mmol)')
ax[0].legend(title='production yield factor')
plt.savefig('results/figures/call11_sal11_mam3/yield_envelope_2D.svg', format='svg')
#ax[1].legend(title='production yield factor')
```
