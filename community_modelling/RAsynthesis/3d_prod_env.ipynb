```python
import pandas as pd
```

```python
df = pd.read_csv("results/prod_env_yields_tri_xyl_glc.csv")
```

```python
df[df["yield_factor"]==0.5]
```

```python
import seaborn as sns

sns.lineplot(data=df, x="time", y="RA", hue="yield_factor")
```

```python
sns.lineplot(data=df, x="time", y="biomass", hue="yield_factor")
```

```python
import pandas as pd
import plotly.express as px

fig = px.scatter_3d(df, x='time', y='biomass', z='RA', color='yield_factor')

marker_size = 4  # Adjust the size of the scatter points as needed

fig.update_traces(marker=dict(size=marker_size))

# Customize the layout and set the width and height of the figure
fig.update_layout(
    width=800,  # Set the width of the figure in pixels
    height=800    # Set the height of the figure in pixels
)

# Show the plot
fig.show()
```

```python
fig.write_image("3d_plot.svg")
```
