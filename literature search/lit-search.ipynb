```python
def askGPT3(title, abstract):

    prompt = 'Does the scientific article: "' + title + '", with the abstract "' + abstract + '", concern product production using a synthetic community or co-culture?'

    response = openai.Completion.create(
        model="text-davinci-002",
        prompt=prompt,
        temperature=0.7,
        max_tokens=193,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )

    return response.choices[0].text 
```

```python
#last in csv fil som dataframe
df = pd.read_csv ('Extracted_articles.csv')
```

```python
answers = []

for i in range(len(df)):
    title = df['Title'][i]
    abstract = df['Abstract'][i]
    answers.append(askGPT3(title, abstract))

df['GPT-3 verdict'] = answers
```
